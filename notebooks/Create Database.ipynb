```python
import mysql.connector
import getpass
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
import pymysql
```

```python
locations = pd.read_csv('locations.csv', sep=';')
locations
```

```python
locations.drop('school', axis='columns', inplace=True)
```

```python
locations.set_index('location_id')
locations.head()
```

```python
saved_password = getpass.getpass()
```

```python
cnx = mysql.connector.connect(user='root', password=saved_password, database='bootcamp_schools_data', port=3306)
```

```python
cnx.is_connected()
```

```python
cursor = cnx.cursor(buffered=True)
```

```python
engine = create_engine("mysql+pymysql://{user}:{pw}@localhost/{db}"
                       .format(user="root",
                               pw=saved_password,
                               db="bootcamp_schools_data"))
```

```python
locations.to_sql('locations', con = engine, if_exists = 'append', chunksize = 1000, index=False)
```

```python
badges = pd.read_csv('badges.csv')
badges.head()
```

```python
badges.columns = ['badge_id', 'school_id', 'school', 'keyword', 'name', 'description']
badges
```

```python
badges.loc[badges['name']=='Job Guarantee', 'badge_id']= 3
```

```python
badges
```

```python
badges.drop('school', axis='columns', inplace=True)
```

```python
badges_school = badges[['badge_id', 'school_id']].copy()
badges_school
```

```python
badges.drop_duplicates(subset='name', inplace=True)
badges.drop('school_id', axis='columns', inplace=True)
badges
```

```python
badges.to_sql('badges', con = engine, if_exists = 'append', chunksize = 1000, index=False)
```

```python
badges_school.to_sql('badges_school', con = engine, if_exists = 'append', chunksize = 1000, index_label='badges_school_id')
```
